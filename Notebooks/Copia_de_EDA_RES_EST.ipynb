# Limpiar el conjunto de datos


Aquí tienes una explicación de cada variable en el conjunto de datos de predicción de salario:

| Variable             | Descripción                                                                  | Tipo de Dato |
| :------------------- | :--------------------------------------------------------------------------- | :----------- |
| `job_title`          | Título del puesto de trabajo.                                               | Categórico   |
| `experience_years`   | Años de experiencia del candidato.                                          | Numérico     |
| `education_level`    | Nivel educativo del candidato (ej. Bachelor, Master, PhD).                  | Categórico   |
| `skills_count`       | Número de habilidades que posee el candidato.                                | Numérico     |
| `industry`           | Industria en la que trabaja el candidato.                                    | Categórico   |
| `company_size`       | Tamaño de la empresa (ej. Small, Medium, Large).                             | Categórico   |
| `location`           | Ubicación geográfica del trabajo.                                            | Categórico   |
| `remote_work`        | Indica si el trabajo es remoto (Yes/No/Hybrid).                              | Categórico   |
| `certifications`     | Número de certificaciones obtenidas por el candidato.                        | Numérico     |
| `salary`             | Salario anual del puesto (variable objetivo).                                | Numérico     |

Aquí tienes un análisis de qué modelos de Machine Learning se pueden aplicar al conjunto de datos de predicción de salario, identificando la variable objetivo y el tipo de aprendizaje:

| Variable Objetivo  | Tipo de Aprendizaje      | Modelos de Machine Learning Sugeridos                                 | Consideraciones                                                                                                                                                             |
| :----------------- | :----------------------- | :-------------------------------------------------------------------- | :-------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `salary`           | Supervisado (Regresión)  | Regresión Lineal, Árboles de Decisión, Random Forest, Gradient Boosting, SVM, Redes Neuronales | Se busca predecir el salario de un puesto de trabajo basándose en sus características. Esta es la tarea principal de este conjunto de datos.                                     |
| `remote_work`      | Supervisado (Clasificación) | Regresión Logística, Árboles de Decisión, Random Forest, SVM, Naive Bayes | Podría ser útil para predecir si un trabajo es remoto o no, basándose en otras características, aunque la naturaleza de esta variable es más de feature engineering.           |
| `education_level`  | Supervisado (Clasificación) | Regresión Logística, Árboles de Decisión, Random Forest, SVM, Naive Bayes | Podría predecirse el nivel educativo requerido para un puesto, si fuera necesario inferir valores faltantes o analizar su relación con otras variables.                      |
| `job_title`        | Supervisado (Clasificación) | Modelos de Clasificación Multiclase, como Random Forest o Gradient Boosting | Predecir el título de trabajo basándose en otras características podría ser una tarea interesante para entender la correlación entre las características y el tipo de trabajo. |
